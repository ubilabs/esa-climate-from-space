Transform cate display settings to yaml style config for xcube (just for convenience)

```python
from cate.conf.defaults import VARIABLE_DISPLAY_SETTINGS
```

```python
VARIABLE_DISPLAY_SETTINGS
```

```python
x = {k:dict(ColorBar=v.get('color_map', 'viridis'), ValueRange=[v.get('display_min', 0), v.get('display_max', 1)]) for k,v in VARIABLE_DISPLAY_SETTINGS.items()}
```

```python
x
```

```python
import yaml
```

```python
yaml.dump(x)
```

```python
with open('./my.yml', 'w') as fp:
    fp.write(yaml.dump(x))
```
